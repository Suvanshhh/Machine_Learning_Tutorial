```python
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv(r'C:\Users\KIIT\MACHINE LEARNING\Insurance_Multi_Reg.csv')
```

```python
df
```

**We have two ways to deal with the missing data:**

**1. Remove the entire section where the value is missing.**

**2. Find the mean, median or mode of the data and replace.**

```python
df.Height.mean()
```

```python
mean_Height=df.Height.mean()
```

```python
df.Height=df.Height.fillna(mean_Height)
```

```python
df
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['Age','Height','Weight']],df[['Premium']])
```

**Formula for linear Regression with multiple variable: Y = m1xX1 + m2xX2 +m3xX3 + c OR Premium= m1xAge + m2xHeight + m3xWeight +c**

```python
reg.coef_
```

```python
reg.intercept_
```

**Since we have to predict the Premium whose age is 27, height is 167.56, and weight is 60 and whose age is 60, height is 165.10, and weight is 80.**

```python
reg.predict([[27,167.56,60]])
```
